```python
import pandas as pd
import tensorflow as tf
```

```python
data=pd.read_csv('housedata.csv')
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
x=data.iloc[:,0:10]
x.drop('SalePrice',axis=1,inplace=True)
```

```python
x=x.values
```

```python
y=data.iloc[:,8:9].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=None)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
classifier=tf.keras.models.Sequential()
```

```python
classifier.add(tf.keras.layers.Flatten())
classifier.add(tf.keras.layers.Dense(units=256,activation='relu'))
classifier.add(tf.keras.layers.Dense(units=512,activation='relu'))
classifier.add(tf.keras.layers.Dense(units=1,activation='linear'))
classifier.compile(optimizer='adam',loss='mean_squared_error')
```

```python
classifier.fit(xtrain,ytrain,batch_size=32,epochs=300)
```

```python
classifier.summary()
```

```python
classifier.evaluate(xtest,ytest)
```

```python
ypred=classifier.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```
